# Распределение Гомпертца

Оно находится в библиотеке scipy.stats (https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.gompertz.html#scipy.stats.gompertz)

## Импортируем библиотеки

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats as sts

from math import sqrt, pi
%matplotlib inline
```

## Генерация выборки объёма 1000

```python
c = 0.947
gompertz_rv = sts.gompertz(c)
date = gompertz_rv.rvs(1000)
```

```python
# Зададим ось x
x = np.linspace(0, 5, 1000) 

# Теоретическая функция распределения:
cdf = gompertz_rv.cdf(x)
plt.plot(x, cdf, label='theoretical CDF')

# Эмпирическая функция распределения (по нашей выборке объёма 1000):
from statsmodels.distributions.empirical_distribution import ECDF
ecdf = ECDF(date)
plt.step(ecdf.x, ecdf.y, label = 'ECDF')

plt.ylabel('f(x)')
plt.xlabel('x')
plt.legend(loc = 'lower right')
```

Благодаря достаточно большому объёму выборки можем убедиться, что эмпирическая функция довольно "близка" к теоретической. Как показана на графике выше

## Построим гистограмму выборки и график теоретической функции плотности распределения нашей случайной величины:

```python
# Гистограмма выборки:
# density альтернатива normed для matlab 3.1
plt.hist(date, density = True)

# Теоретическая плотность распределения
x = np.linspace(-1, 5, 1000) # start = -1, stop = 5, num = 1000
pdf = gompertz_rv.pdf(x)
plt.plot(x, pdf, label = 'Theoretical pdf', color = 'black')

plt.legend()
plt.grid()

plt.ylabel('f(x)')
plt.xlabel('x')
```

# Применение центральной предельной теоремы
Для n = (2, 5, 10, 25, 45) cгенерировать 1000 выборок объёма n, для каждой считать выборочное среднее и строить гистограммы распределений выборочных средних для каждого n, а поверх гистограмм строить графики плотности вероятности для нормальных распределений.

```python
n_list = [2, 5, 10, 25, 45] # запишем n в лист


for n in n_list:
    sample_list = []
    for i in range(0, 1001):
        date_sample = gompertz_rv.rvs(n)
        cur_mean = sum(date_sample) / float(n)
        sample_list.append(cur_mean)

    # density альтернатива normed для matlab 3.1
    plt.hist(sample_list, bins=20, density=True)

    # Мат. ожидание и дисперсия распределния Гомпертца
    th_mean, th_disp, skew, kurt = sts.gompertz.stats(c, moments='mvsk')
    
    # Строим график плотности вероятности нормального распределения по ЦПТ
    norm_rv = sts.norm(th_mean, sqrt(th_disp / n))
    x = np.linspace(0, 3, 1000)
    pdf = norm_rv.pdf(x)
    plt.plot(x, pdf, label = 'Current norm pdf', alpha = 0.5, color = 'black')

    plt.title('n = {}'.format(n))
    plt.legend(loc = "upper right")
    plt.ylabel('f(x)')
    plt.xlabel('x')
    plt.show()
```
