```python
import pandas as pd
import re
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import scipy
import seaborn as sns
%matplotlib inline
%config IPCompleter.greedy=True
```

### Геренируем выборку  для показательного распредления и строим гистограмму

```python
from scipy.stats import expon
rv = expon()

sample = rv.rvs(1000)
#sample
```

```python
x = np.linspace(0,7,100)
cdf = rv.pdf(x)
plt.plot(x, cdf, label='theoretical CDF')



plt.legend(loc='upper left')

plt.hist(sample, normed=True)
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
```

### Генерируем 1000 выборок  для n=5

```python
L=1
n=5
rv = expon(scale=1/L)
sample = rv.rvs([1000,n])

average=list()
for i in range(1000):
    average.append(np.sum(sample[i])/n)
```

###  Строим гистограмму и плотность соответствующего нормального распределения 

```python
x = np.linspace(0,3,100)
norm = sts.norm(loc=1/L,scale=np.sqrt((1/L)**2/n))

plt.hist(average, normed=True)
plt.plot(x,norm.pdf(x), label='norm_rv')

plt.legend(loc='upper left')

plt.ylabel('fraction of samples')
plt.xlabel('$x$')
```
